## Cleaning

```python
import re
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
```

```python
print(dirty)
```

```python
clean=re.sub(r'https:.*','',dirty) # remove URLs
clean=re.sub(r'@[a-zA-z0-9]*','',clean) # remove @WORD
clean=re.sub(r'#[a-zA-z0-9]*','',clean) # remove #hashtags
clean=re.sub(r' +',' ',clean) # remove multiple spaces
clean=re.sub(r'\n+','\n',clean) # remove multiple new lines
clean=emoji_pattern.sub(r'', clean) # remove emojis
# clean=re.sub(r'[^a-zA-Z0-9_:!?\.,\n ]+', '', clean)
print(clean)
```

## Chunking

```python
from spacy.matcher import Matcher
```

```python
print("Predefined noun chunks:")
for nc in doc.noun_chunks:
    print("  ",nc)
```

```python
matcher = Matcher(nlp.vocab)

nouns = [{"POS": "NOUN", "OP":"+"}]
nphrases = [{"POS": {"IN":["NOUN","ADJ"]}, "OP":"*"},{"POS": "NOUN", "OP":"+"}]
passiv=[{"LEMMA":"be"},{"POS":"VERB"}]

matcher.add("nouns", None, nouns)
matcher.add("nphrases", None, nphrases)
matcher.add("passiv", None, passiv)
matches = matcher(doc)
for match_id, start, end in matches:
    string_id = nlp.vocab.strings[match_id]  # Get string representation
    span = doc[start:end]  # The matched span
    print(string_id, start, end, span.text)
```

## Nested chunking

```python
grammar = """NP: {<DT>?<JJ|JJR|NN|NNP|NNPS>*<NN|NNS|NNP|NNPS>+}
             CP: {<NP><CC><NP>}
             BP: {<NP|CP>(<IN><NP|CP>)+}
             PP: {<NP|CP><.*><VBN>}
             NM: {<CD><NP>}
             VP: {<RB>?<V.*>+}
          """
parser = nltk.RegexpParser(grammar)
trees=[]
for sent,i in zip(doc.sents,range(len(list(doc.sents)))):
    parsed = [(t.orth_,t.tag_) for t in sent]
    result = parser.parse(parsed)
    print(i,sent)
    print(result)
    trees.append(result)
```

```python
from IPython.display import display
display(trees[6]) # 3,6
```

## Chinking

```python
grammar = """
  NP:
    {<.*>+}          # Chunk everything
    }<V.*|RB|WP|WDT|TO|\.>+{      # Chink sequences of VBD and IN
  """
parser = nltk.RegexpParser(grammar)
trees2=[]
for sent,i in zip(doc.sents,range(len(list(doc.sents)))):
    parsed = [(t.orth_,t.tag_) for t in sent]
    result = parser.parse(parsed)
    print(i,sent)
    print(result)
    trees2.append(result)
```

## Word Embedding

```python
term = "coronavirus"
```

```python
nlp(term)[0].vector
```

```python
nlp(term)[0].vector.shape
```

```python
for sent in list(doc.sents)[:2]:
    print(sent)
    for t in sent:
        if not t.has_vector or t.is_stop:
            continue
        print('  ',t,"->","similarity to %s:"%term,t.similarity(nlp(term)[0]))
```

```python
sim_dict = {}
for t in doc:
    if not t.has_vector:
        continue
    sim_dict[t.text]=t.similarity(nlp(term)[0])
```

```python
sorted(sim_dict.items(),key=lambda x:x[1],reverse=True)
```

## Keywords extraction

```python
from spacy.lang.en.stop_words import STOP_WORDS
```

```python
STOP_WORDS
```

```python
print("TOP WORDS")
topwords = {}
for t in doc:
    if t.is_stop or t.is_punct or t.text==" ":
        continue
    if t.text in topwords:
        topwords[t.lemma_]+=1
    else:
        topwords[t.lemma_]=1
sorted(topwords.items(),key=lambda x:x[1],reverse=True)
```

```python
df = {}
for key in topwords:
    cnt = 0
    df[key]=0
    for sent in doc.sents:
        if key in [t.lemma_ for t in sent]:
            df[key]+=1
df
```

```python
import math
idf = {}
for key in df:
    idf[key] = math.log(len(list(doc.sents))/df[key])
idf
```

```python
tfidf={}
for key in idf:
    tfidf[key] = topwords[key]*idf[key]
sorted(tfidf.items(),key=lambda x:x[1],reverse=True)
```
